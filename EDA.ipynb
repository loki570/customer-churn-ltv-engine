# Customer Churn Prediction - EDA

Full Exploratory Data Analysis using Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("../data/Customer Churn Prediction & Lifetime Value(4).xlsx")
df.head()
```

```python
df.shape
df.info()
df.describe(include='all')
```

```python
# Missing values
df.isnull().sum()
```

```python
# Data cleaning
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
df.isnull().sum()
```

```python
# Churn distribution
df['Churn'].value_counts()

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Churn')
plt.title('Customer Churn Distribution')
plt.show()
```

```python
# Contract vs churn
pd.crosstab(df['Contract'], df['Churn'], normalize='index')*100

sns.countplot(data=df, x='Contract', hue='Churn')
plt.xticks(rotation=45)
plt.title('Contract vs Churn')
plt.show()
```

```python
# Tenure analysis

df.groupby('Churn')['tenure'].mean()

sns.boxplot(data=df, x='Churn', y='tenure')
plt.title('Tenure vs Churn')
plt.show()
```

```python
# Monthly Charges

df.groupby('Churn')['MonthlyCharges'].mean()

sns.boxplot(data=df, x='Churn', y='MonthlyCharges')
plt.title('Monthly Charges vs Churn')
plt.show()
```

```python
# Total Charges

df.groupby('Churn')['TotalCharges'].mean()

sns.boxplot(data=df, x='Churn', y='TotalCharges')
plt.title('Total Charges vs Churn')
plt.show()
```

```python
# Correlation

numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## Business Findings

- Month-to-month contracts have highest churn
- Long tenure customers are more loyal
- High monthly charges increase churn risk
- Short tenure + high charges = high risk customers
