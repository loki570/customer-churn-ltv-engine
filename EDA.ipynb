```python
# Outlier Detection using IQR and Z-score methods

numerical_cols = df.select_dtypes(include=[np.number]).columns

# Method 1: IQR (Interquartile Range)
print("=== Outlier Detection using IQR Method ===")
outliers_iqr = {}
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outliers_iqr[col] = len(outliers)
    print(f"{col}: {len(outliers)} outliers (bounds: {lower_bound:.2f} - {upper_bound:.2f})")

# Method 2: Z-score
print("\n=== Outlier Detection using Z-score Method ===")
outliers_zscore = {}
z_scores = np.abs(stats.zscore(df[numerical_cols].dropna()))
threshold = 3

for idx, col in enumerate(numerical_cols):
    outlier_count = (z_scores[:, idx] > threshold).sum()
    outliers_zscore[col] = outlier_count
    print(f"{col}: {outlier_count} outliers (Z-score > {threshold})")

# Visualize Outliers
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].bar(outliers_iqr.keys(), outliers_iqr.values(), color='skyblue')
axes[0].set_title('Outliers Count - IQR Method')
axes[0].set_ylabel('Number of Outliers')
axes[0].tick_params(axis='x', rotation=45)

axes[1].bar(outliers_zscore.keys(), outliers_zscore.values(), color='lightcoral')
axes[1].set_title('Outliers Count - Z-score Method')
axes[1].set_ylabel('Number of Outliers')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

```
